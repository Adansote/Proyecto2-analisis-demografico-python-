Con apoyo de la IA se llega a otro resultado de manera vista mas estadisticamente
"El Sesgo de EU (Cantidad Absoluta): Estados Unidos tiene, por naturaleza, la mayor cantidad de ganadores de altos ingresos en el dataset porque es el país con el mayor número de registros. Sin embargo, esta cifra absoluta no dice nada sobre la probabilidad de que un residente de EE. UU. gane más de $50K.

La Fuerza de Irán (Tasa de Éxito): Irán tiene una población mucho menor. Su alta tasa de éxito (porcentaje) significa que de cada 100 iraníes en el dataset, un mayor número proporcional gana más de $50K en comparación con 100 estadounidenses.

Conclusión: La tasa de éxito de Irán es superior a la de Estados Unidos, porque su porcentaje de ganadores sobre su total de población muestreada es mayor, incluso si el número absoluto de ganadores es mucho menor."

```python
# 1. Agrupar por país y calcular el total de registros por país (el DENOMINADOR)
total_por_pais = df['native.country'].value_counts()
```

```python
total_por_pais
```

```python
# 2. Agrupar por país y calcular el total de ganadores > $50K (el NUMERADOR)
# Convertimos el booleano (income == '>50K') a entero (True=1, False=0) y sumamos
ganadores_por_pais = df.groupby('native.country')['income'].apply(lambda x: (x == '>50K').sum())
```

```python
ganadores_por_pais
```

```python
# 3. Calcular el Porcentaje: Dividir los ganadores por el total de personas en ese país
porcentaje_ganador = (ganadores_por_pais / total_por_pais) * 100
```

```python
porcentaje_ganador
```

```python
# 4. Encontrar el país con el porcentaje MÁXIMO
pais_max_porcentaje = porcentaje_ganador.idxmax()
porcentaje_maximo = porcentaje_ganador.max()
```

```python
# --- Mostrar Resultados ---

print(f"Total de países analizados: {len(total_por_pais)}")
print("-" * 50)
print(f"🏆 País con el MAYOR PORCENTAJE de personas que ganan > $50K: {pais_max_porcentaje}")
print(f"✅ Porcentaje Máximo (dentro del país): {porcentaje_maximo:.2f}%")
```
